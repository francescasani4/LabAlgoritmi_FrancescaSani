## Cenni teorici <a id="teoria">

### Problema dell'ordinamento <a id="problema">
Data una sequenza di *n* numeri potremmo avere la necessità di volerla ordinare secondo determinati criteri, è possibile formalizzare il *problema dell'ordinamento* come segue: <br>
**Input** : una sequenza di *n* numeri $ A = <a_1, a_2, ..., a_n>$ <br>
**Output** : una permutazione $A' = <a'_1, a'_2, ..., a'_n>$ di $A$ tale che $a'_1 \leq a'_2 \leq ... \leq a'_n$.

Esistoni vari algoritmi di ordinameno che permettono di "risolvere" questo problema. La scelta di quale algoritmo utilizzare dipenderà dal contesto in cui dovrà operare, poiché differiscono tra loro per velocità di esecuzione e 
complessità computazionale.

---
### Selection-Sort <a id="Selection-Sort">
Il **Selection-Sort**, o ordinamento per selezione, è un algoritmo di ordinamento il quale seleziona l'elemento più piccolo da un array non ordinato ad ogni iterazione e posiziona tale elemento all'inizio dell'array non ordinato. Questo tipo di ordinamento è un ordinamento di tipo crescente.

Inizialmente, il primo elemento viene impostato come **minimo** e lo confronta con quello successivo, che se minore diventa il nuovo minimo, altrimenti procede con il successivo, questo lo fa fino ad arrivare all'ultimo elemento. <br>
Dopo ogni iterazione il *minimo* individuato viene scambiato con il primo elemento da analizzare. <br>
Questo procedimento viene eseguito finché tutti gli elementi presenti nell'array non si trovano nella posizione corretta, ovvero finché non resta un solo elemento che sarà sicuramente maggiore rispetto a gli elementi di sinistra contenuti nell'array.

L'algoritmo Selection-Sort è un algoritmo di ordinamento che ordina *sul posto*.

### Complessità <a id="complessitàS">
La complessità di questo algoritmo è la solita in tutti i casi, poiché ad ogni passo devo trovare il minimo e metterlo nella posizione corretta. L'elemento minimo non viene individuato finché tutto l'array non è stato esplorato. <br> 
Quindi sia nel caso peggiore (array ordinato inversamente), sia nel caso migliore (array ordinato), sia in quello medio (array in ordine casuale) la complessità è di $O(n^2)$.

| Caso peggiore | Caso migliore | Caso medio |
|:---:|:---:|:---:|
| $O(n^2)$ | $O(n^2)$ | $O(n^2)$ |

Questo algoritmo non è molto efficiente, ma è un'ottima soluzione quando si deve ordinare un quantitativo non troppo elevato di elementi.

### Correttezza <a id="correttezzaS">
L'algoritmo Selection-Sort è un algoritmo iterativo, pertanto è possibile dimostrare la sua **correttezza** attraverso un invariante di ciclo; è possibile affermare la correttezza di questo algoritmo, poiché qualsiasi input che gli viene fornito termina sempre con l'output corretto.

### Stabilità <a id="stabilitàS">
Il Selection-Sort è un algoritmo **non stabile**, poiché chiavi che hanno lo stesso valore possono essere invertite e quindi non apparire nello stesso ordine di ingresso. 

### Codice <a id="codiceS">

```python
# Implementazione dell'algoritmo Selection-Sort
def SelectionSort(A: list):
    n = len(A)

    for i in range(n):
        min_idx = i

        for j in range(i + 1, n):
            if A[j] < A[min_idx]:
                min_idx = j
        tmp = A[i]
        A[i] = A[min_idx]
        A[min_idx] = tmp
```

### Quick-Sort <a id="Quick-Sort">
Il **Quick-Sort** è un algoritmo basato sul paradigma *Divide-at-Impera*, che sceglie un elemento come pivot e partiziona l'array dato in base al pivot selezionato, posizionandolo nella sua posizione corretta. 

Per ordinare un array della forma $A[p ... r]$: <br>

**Divide**: Partizionare l'array $A[p ... r]$ in due sottoarray $A[p ... q-1]$ e $A[q+1 ... r]$ tali che ogni elemento in $A[p ... q-1]$ sia minore uguale a $A[q]$  e ogni elemento in $A[q+1 ... r]$ sia maggiore uguale a $A[q]$. <br>
**Impera**: Ordinare i due sottoarray attraverso chiamate ricorsive a *QUICK-SORT*. <br>
**Combina**: I sottoarray sono ordinati sul posto, l'intero array $A[p ... r]$ è ordinato.

L'algoritmo Quick-Sort è un algoritmo di ordinamento che ordina *sul posto*.

### Complessità <a id="complessitàQ">
La complessità di questo algoritmo si differenza tra caso peggiore e caso migliore e caso medio.<br>
    
* *Caso migliore*, ovvero quando il pivot viene scelto im modo tale che divide sempre l'array a metà e quindi l'algoritmo crea delle partizioni bilanciate, la sua complessità è di $O(n\  lg\  (n))$.<br>
* *Caso peggiore*, ovvero quando il pivot viene scelto sempre in modo tale che sia sempre o l'elemento più piccolo o l'elemento più grande, creando così delle partizioni totalmente sbilanciate, pertanto la sua complesità è di $O(n^2)$.<br>
* *Caso medio* la sua complessità è di $O(n\ lg\ (n))$.

| Caso peggiore | Caso migliore | Caso medio |
|:---:|:---:|:---:|
| $O(n^2)$ | $O(n\ lg\ (n))$ | $O(n\ lg\ (n))$ |

Questo algoritmo è efficiente se si riesce ad evitare il verificarsi del caso peggiore, per esempio scegliendo il pivot in modo casuale. Inoltre, è vantaggioso in caso di un quantitativo elevato di elementi da ordinare.

### Correttezza <a id="correttezzaQ">
Il Quick-Sort è un algoritmo basato sul partizionamento, per verificare la **correttezza** della partizione, essendo iterativa, è possibile farlo attraverso un invariante di ciclo. Inoltre, come per il Selection-Sort, è possibile affermare che tale algoritmo genera sempre l'output corretto per qualsiasi input dato.

### Stabilità <a id="stabilitàS">
Il Quick-Sort, come il Selection-Sort, è un algoritmo **non stabile**, poiché chiavi che hanno lo stesso valore possono essere invertite e quindi non apparire nello stesso ordine di ingresso.

### Codice <a id="codiceQ">

```python
# Implementazione dell'algoritmo Partition per Quick-Sort
def Partition(A, p, r):
    pivot = A[r]
    i = p -1
    for j in range(p, r):
        if A[j] <= pivot:
            i = i + 1
            tmp= A[i]
            A[i] = A[j]
            A[j] = tmp

    tmp = A[i+1]
    A[i+1] = A[r]
    A[r] = tmp
    return i + 1
```

```python
# Implementazione dell'algoritmo ricorsivo Quick-Sort con chiamate Ricorsive
def QuickSortRecursive(A, p, r):
    if p < r:
        q = Partition(A, p, r)
        QuickSortRecursive(A, p, q-1)
        QuickSortRecursive(A, q+1, r)
```

```python
# Implementazione della prima chiamata di Quick-Sort
def QuickSort(A: list):
    r = len(A) - 1
    QuickSortRecursive(A, 0, r)
```

### Randomized Quick-Sort <a id="randomQuickSort">
    
Analizzando il caso medio dell'algoritmo Quick-Sort, è stato possibile notare che tutte le permutazioni di ingresso sono ugualmente probabili, ma questo non sempre è verificato. <br>
    
Posso rendere l'algoritmo Quick-Sort **randomizzato** permutando casualmente l'input, oppure scegliendo un elemento a caso come pivot, questo mi permette di ottenere buone prestazioni su tutti gli input. <br>
    
In questo caso andremo ad analizzare la tecnica del *campionamento casuale*, ovvero invece di usare sempre $A[r]$ come pivot, viene selezionato un elemento selto a caso nel sottoarray $A[p...r]$. <br> Dato che l'elemento pivot è scelto casualmente, ci aspettiamo che la divisione dell'array in input sia bilanciata.

#### Codice <a id="randomCodice">

```python
import random

def randomizedPartition(A, p, r):
    i = random.choice([p, r])
    A[r], A[i] = A[i], A[r]
    return Partition(A, p, r)
```

```python
def randomizedQuickSortRecursive(A, p, r):
    if p < r:
        q = randomizedPartition(A, p, r)
        randomizedQuickSortRecursive(A, p, q - 1)
        randomizedQuickSortRecursive(A, q + 1, r)
```

```python
def randomizedQuickSort(A: list):
    r = len(A) - 1
    randomizedQuickSortRecursive(A, 0, r)
```

### Quick-Sort con pivot centrale <a id="bestQuickSort">
Come è stato detto in precedenza, per avere il caso migliore è necessario che il pivot divida sempre a metà l'array da ordinare, in modo da avere delle partizioni bilanciate. <br>
Per fare questo è stato definito il seguente algoritmo che verrà in seguito utilizzato per verificare il caso migliore. <br>
    
Questo algoritmo sceglie sempre come pivot l'elemento centrale.

#### Codice <a id="bestCodice">

```python
def bestPartition(A, p, r):
    i = int(r/2)
    A[r], A[i] = A[i], A[r]
    return Partition(A, p, r)
```

```python
def bestQuickSortRecursive(A, p, r):
    if p < r:
        q = bestPartition(A, p, r)
        bestQuickSortRecursive(A, p, q - 1)
        bestQuickSortRecursive(A, q + 1, r)
```

```python
def bestQuickSort(A: list):
    r = len(A) - 1
    bestQuickSortRecursive(A, 0, r)
```

## Descrizione dei Test <a id="descTest">
Per poter testare le prestazioni dei due diversi algoritmi di ordinamento, sarà necerrario generare una lista di *n* valori e misurare i rispettivi tempi di esecuzione. <br>
    
Per eseguire e semplificare questi test vengono presi in considerazione soltanto valori numerici interi. <br>
Gli array che verranno generati, saranno tali da poter testare al meglio l'algoritmo Quick-Sort,  il quale ha una complessità diversa nei tre casi, a differenza del Selection-Sort che ha sempre una complessità di $O(n^2)$. <br>

I test di ordinamento verranno eseguiti su un insieme di *n* valori, con *n* che varia da 1 a 3000. <br>
Per ogni test, verranno eseguite più prove, generando input diversi, e verrà inoltre calcolato un valore medio per ciascun gruppo di prove effettuato.

### Generazione dati per i Test <a id="genDataTest">

Definiamo tre funzioni che permettono di generare array con *n* valori:
* La prima funzione, <code>sortedArray(n)</code>, genera array di *n* elementi ordinati in modo crescente;
* La seconda funzione, <code>reversedArray(n)</code>, genera array di *n* elementi ordinati in modo decrescente;
* La terza funzione, <code>randomArray(n)</code>, genera array di *n* valori casuali compresi tra 1 e un valore massimo (max).

```python
import numpy as np

max = 500

def sortedArray(n):
    return list(range(n))

def reversedArray(n):
    return list(range(n-1, -1, -1))

def randomArray(n):
    return np.random.randint(0, max+1, n).tolist()
```

### Esecuzione dei Test <a id="esecTest">

Per eseguire i test è stato definito un numero totale di iterazioni, *n*. Ad ogni iterazione viene generato un array composto da **i** elementi, il quale sarà ordinato utilizzando i due agoritmi in esame. <br>

La misurazione dei tempi di esecuzione è stata fatta attraverso il modulo **timeit**, in particolare attraverso la funzione <code>timeit.timeit()</code>. L'ordinamento è stato ripetuto un numero di volte pari a *nrIterazioniTest*, in modo da avere una misurazione più adeguata, e in seguito è stata calcolata la media dei tempi ottenuti. <br>

Da notare che per l'algoritmo Quick-Sort non è sempre stata utilizzata la solita funzione, poiché non avrebbe verificato in modo adeguato le tre casistiche, pertanto sono state utilizzate le seguenti funzioni (definite in precedenza):
* <code>bestQuickSort</code> per verificare il caso migliore con in input un array ordinato;
* <code>QuickSort</code> per verificare sia il caso peggiore con in input un array ordinato inversamente sia il caso medio con in input un array casuale;
* <code>randomizedQuickSort</code> per verificare il caso medio con in input un array casuale.

Invece, per quanto riguarda il Selection-Sort, è sempre stata utilizzata la funzione <code>SelectionSort</code> con in input array diversi in base al caso in esame.

```python
import timeit

n = 3000
j = 150
nrIterazioniTest = 100

selcetionSortBest = []
quickSortBest = []

selcetionSortWorst = []
quickSortWorst = []

selcetionSortAverage = []
quickSortAverage = []

selcetionSortRandomized = []
quickSortRandomized = []


def measureTime(function, args):
    """ Ritorna il tempo di esecuzione medio (in millisecondi) della funzione data"""
    return timeit.timeit(stmt=lambda : function(args), number=nrIterazioniTest) / nrIterazioniTest * 1000
```

```python
# Caso migliore
for i in range(1, n, j):
    arrayTestS = sortedArray(i)
    arrayTestQ = arrayTestS

    selcetionSortBest.append(measureTime(SelectionSort, arrayTestS))

    quickSortBest.append(measureTime(bestQuickSort, arrayTestQ))
    
print("Eseguito")
```

```python
# Caso peggiore
for i in range(1, n, j):
    arrayTestS = reversedArray(i)
    arrayTestQ = arrayTestS

    selcetionSortWorst.append(measureTime(SelectionSort, arrayTestS))

    quickSortWorst.append(measureTime(QuickSort, arrayTestQ))

print("Eseguito")
```

```python
# Caso medio
for i in range(1, n, j):
    arrayTestS = randomArray(i)
    arrayTestQ = arrayTestS

    selcetionSortAverage.append(measureTime(SelectionSort, arrayTestS))

    quickSortAverage.append(measureTime(QuickSort, arrayTestQ))
    
    
print("Eseguito")
```

```python
# Quick-Sort randomized caso medio
for i in range(1, n, j):
    arrayTestS = randomArray(i)
    arrayTestQ = arrayTestS

    selcetionSortRandomized.append(measureTime(SelectionSort, arrayTestS))

    quickSortRandomized.append(measureTime(randomizedQuickSort, arrayTestQ))
    
print("Eseguito")
```

### Generazione dei grafici di Test <a id="graficiTest">
    
Attraverso l'uso del modulo <code>matplotlib</code> sono stati generati i grafici sottostanti. <br>

I grafici fanno riferimento rispettivamente al *caso migliore* (array con valori ordinati in modo crescente), al *caso peggiore* (array con valori ordinati in modo decrescente) e al *caso medio* (array con valori causali).

```python
import matplotlib.pyplot as plt

def drawPlots(leftData, rightData, plotTitle: str = None):
    x = np.linspace(1, n, len(leftData))
    fig, (left, right) = plt.subplots(1, 2, figsize=(15, 5))

    #Selection-Sort
    left.plot(x, leftData, color="green")
    left.set_title("Selection-Sort")
    left.set_xlabel("Dimensioni della lista")
    left.set_ylabel("Tempo di esecuzione (ms)")

    # Quick-Sort
    right.plot(x, rightData, color="blue")
    right.set_title("Quick-Sort")
    right.set_xlabel("Dimensioni della lista")
    right.set_ylabel("Tempo di esecuzione (ms)")

    if plotTitle:
        fig.suptitle(plotTitle, fontsize=16)

def plots(leftData, rightData, plotTitle: str = None):
    x = np.linspace(1, n, len(leftData))
    fig, ax = plt.subplots(1, figsize=(15, 5))

    #Selection-Sort
    plt.plot(x, leftData, label="Selection-Sort", color="green")

    # Quick-Sort
    plt.plot(x, rightData, label="Quick-Sort", color="blue")
    
    plt.xlabel("Dimensioni dell'array")
    plt.ylabel("Tempo di esecuzione (ms)")
    
    plt.legend()
    plt.show

    if plotTitle:
        fig.suptitle(plotTitle, fontsize=16)
```

```python
drawPlots(selcetionSortBest, quickSortBest, "Caso Migliore (Figura 3)")
plots(selcetionSortBest, quickSortBest, "Caso Migliore (Figura 4)")
```

```python
drawPlots(selcetionSortWorst, quickSortWorst, "Caso Peggiore (Figura 5)")
plots(selcetionSortWorst, quickSortWorst, "Caso Peggiore (Figura 6)")
```

```python
drawPlots(selcetionSortAverage, quickSortAverage, "Caso Medio (Figura 7)")
plots(selcetionSortAverage, quickSortAverage, "Caso Medio (Figura 8)")
```

```python
drawPlots(selcetionSortRandomized, quickSortRandomized, "Caso Medio con Randomized (Figura 9)")
plots(selcetionSortRandomized, quickSortRandomized, "Caso Medio con Randomized (Figura 10)")
```

### Generazione delle tabelle di Test <a id="tabelleTest">

Le tabelle sottostanti, che rappresentano i tempi di esecuzione degli algoritmi in esame, sono state generate attraverso l'uso del modulo <code>numpy</code>.

```python
bestData = [
    [i for i in range(1, n, j)],
    ["{:.3e}".format(val) for val in selcetionSortBest],
    ["{:.3e}".format(val) for val in quickSortBest]
]

worstData = [
    [i for i in range(1, n, j)],
    ["{:.3e}".format(val) for val in selcetionSortWorst],
    ["{:.3e}".format(val) for val in  quickSortWorst]
]

averageData = [
    [i for i in range(1, n, j)],
    ["{:.3e}".format(val) for val in selcetionSortAverage],
    ["{:.3e}".format(val) for val in quickSortAverage]
]

randomizedData = [
    [i for i in range(1, n, j)],
    ["{:.3e}".format(val) for val in selcetionSortRandomized],
    ["{:.3e}".format(val) for val in quickSortRandomized]
]
 
def tables(columns: list, headers: tuple, title: str):
    fig, ax = plt.subplots(figsize=(10, 7)) #serve per sistemare il titolo da rivedere
    plt.title(title)
    
    data = np.stack(tuple(columns), axis=1) 

    ax.axis('off')
    table = ax.table(cellText=data, colLabels=headers, loc='center', cellLoc='center')
    table.auto_set_column_width(col=list(range(len(columns))))
    table.scale(1, 1.5)

    cell_colors = {
        cell: ("#e0ebd4", {"weight": "bold"})
        if table[cell].get_text().get_text() in headers
        else ("#e0ebd4", {})
        for cell in table._cells
        if cell[0] % 2 == 0
    }
    for cell, (color, text_props) in cell_colors.items():
        table[cell].set_facecolor(color)
        table[cell].set_text_props(**text_props)
        
    plt.show()
```

```python
tables(bestData,("Nr elementi", "Selection-Sort", "Quick-Sort"), "Caso migliore (Figura 11)")
```

```python
tables(worstData,("Nr elementi", "Selection-Sort", "Quick-Sort"), "Caso peggiore (Figura 12)")
```

```python
tables(averageData,("Nr elementi", "Selection-Sort", "Quick-Sort"), "Caso medio (Figura 13)")
```

```python
tables(randomizedData,("Nr elementi", "Selection-Sort", "Quick-Sort"), "Caso medio con Randomized (Figura 14)")
```

### Caso migliore: array con valori ordinati  <a id="casoMigliore">
    
Il caso migliore per Selection-Sort si ha quando i dati in input sono ordinati, la sua complessità è quadratica.<br>
Invece, per Quick-Sort il caso migliore si ha quando la partizione è bilanciata, ovvero quando si ha una array ordinato e viene scelto come pivot sempre la metà, la sua complessità è logaritmica.<br> <br> 
Attraverso i grafici in *figura 3 e 4* è possibile verificare i due diversi andamenti, in particolare nel grafico in basso (*figura 4*) e nella tabella in *figura 11* è possibile osservare che i tempi di esecuzione dell’algoritmo Quick-Sort sono inferiori a quelli di Selection-Sort. Da questi test è possibile verificare ciò che è stato affermato in precedenza nei cenni teorici. 

### Caso peggiore: array con valori ordinati inversamente <a id="casoPeggiore">
    
Il caso peggiore per Selection-Sort si ha quando i dati in input sono ordinati inversamente, la sua complessità è quadratica. In questo caso anche Quick-Sort ha una complessità quadratica.
Questo loro andamento è osservabile dai grafici in *figura 5*. <br> <br>
Attraverso la tabella in *figura 12* e il grafico in *figura 6* è possibile vedere che, nonostante l’andamento sia quadratico per entrambi gli algoritmi, i tempi di esecuzione di Quick-Sort rispetto a quelli di Selection-Sort sono maggiori.

### Caso medio: array con valori casuali <a id="casoMedio">

Il caso medio per il Selection-Sort si verifica quando si ha in input un array con valori casuali e la sua complessità è quadratica, come è osservabile nel grafico di sinistra in *figura 7*. <br> Per verificare il caso medio di Quick-Sort sono stati eseguiti due test: 
1. Nel primo caso di test (*figura 7*) sono stati considerati in ingresso valori casuali ed è stato utilizzato l’algoritmo che sceglie come pivot sempre l’ultimo elemento; 
2. Nel secondo caso di test (*figura 9*) oltre a considerare valori casuali in ingresso è stato utilizzato l’algoritmo che sceglie casualmente il pivot.
    
Grazie ai grafici in *figura 8 e 10* e alle tabelle in *figura 13 e 14* è osservabile che Quick-Sort ha tempi di esecuzione inferiori rispetto a Selection-Sort, quando si esegue il secondo caso di test; invece ha tempi di esecuzione maggiori rispetto a Selection-Sort, quando si esegue il primo caso di test. Inoltre, dalla tabella in *figura 14* è osservabile che quando si ha pochi valori nell’array i tempi di esecuzione di Selection-Sort sono inferiori a quelli di Quick-Sort, ma al crescere della quantità di dati nell’array (quando si ha circa più di 1000 elementi) la situazione si ribalta, questo comportamento è osservabile anche dal grafico in *figura 10*.
